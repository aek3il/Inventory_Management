```python
import json
fd = open("inventory.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
records
```

```python
prod = str(input("Enter Product ID: "))
name = str(input("Enter Name: "))
price = int(input("Enter Price: "))
quant = int(input("Enter Quantity: "))
cate = str(input("Enter Category: "))
rate = int(input("Enter Ratings: "))

records[prod] = {'name': name, 'price': price, 'quantity': quant, 'category': cate, 'rating': rate}

js = json.dumps(records)

fd = open("inventory.json",'w')
fd.write(js)
fd.close()
```
